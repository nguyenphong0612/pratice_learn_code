```python
from tkinter import *
# Khởi tạo cửa sổ màn hình
calculator = Tk()
calculator.title("Calculator")
calculator.geometry("350x100")
# định nghĩa hàm tính toán
def calculate():
    global result
    formula = formula_field.get()
    formula_field.delete(0, END)
    result = eval(formula)
    #print(result)
    formula_field.insert(0, str(result))

def in_char(char):
    #formula_field.delete(0, END)
    formula_field.insert(END, char)   
    if char == "AC":
        formula_field.delete(0, END)
    #formula_field.delete(0, END)
# Tạo các widget
# entry
formula_field = Entry(calculator, width=45)

# button
num0_btn = Button(calculator, text="0", width=5, command= lambda: in_char(0))
num1_btn = Button(calculator, text="1", width=5, command= lambda: in_char(1))
num2_btn = Button(calculator, text="2", width=5, command= lambda: in_char(2))
num3_btn = Button(calculator, text="3", width=5, command= lambda: in_char(3))
num4_btn = Button(calculator, text="4", width=5, command= lambda: in_char(4))
num5_btn = Button(calculator, text="5", width=5, command= lambda: in_char(5))
num6_btn = Button(calculator, text="6", width=5, command= lambda: in_char(6))
num7_btn = Button(calculator, text="7", width=5, command= lambda: in_char(7))
num8_btn = Button(calculator, text="8", width=5, command= lambda: in_char(8))
num9_btn = Button(calculator, text="9", width=5, command= lambda: in_char(9))
sum_btn = Button(calculator, text="+", width=5, command= lambda: in_char("+"))
sub_btn = Button(calculator, text="-", width=5, command= lambda: in_char("-"))
mul_btn = Button(calculator, text="*", width=5, command= lambda: in_char("*"))
div_btn = Button(calculator, text="/", width=5, command= lambda: in_char("/"))
dots_btn = Button(calculator, text=".", width=5, command= lambda: in_char("."))
cal_btn = Button(calculator, text="calculate", width=7, command=calculate)
AC_btn = Button(calculator, text="AC", width=5, command= lambda: in_char("AC"))
# dán các widget lên màn hình
formula_field.grid(row=0, column=0, columnspan=5)
num0_btn.grid(row=1, column=0)
num1_btn.grid(row=1, column=1)
num2_btn.grid(row=1, column=2)
num3_btn.grid(row=1, column=3)
num4_btn.grid(row=1, column=4)
num5_btn.grid(row=2, column=0)
num6_btn.grid(row=2, column=1)
num7_btn.grid(row=2, column=2)
num8_btn.grid(row=2,column=3)
num9_btn.grid(row=2, column=4)
sum_btn.grid(row=3, column=0)
sub_btn.grid(row=3, column=1)
mul_btn.grid(row=3, column=2)
div_btn.grid(row=3, column=3)
dots_btn.grid(row=3, column=4)
cal_btn.grid(row=1, column=5)
AC_btn.grid(row=2, column=4)
calculator.mainloop()
```
